```python
# 6 . how many of them have not logged there activity find out in terms of number of ids 
pd.read_sql('select count(*) as not_logged from (select Id, sum(LoggedActivitiesDistance) as lad from `fitbit data` group by Id having lad<1) as t',mydb)
```

```python
# 8 . Explore over an internet that how much calories burn is required for a healthy person and find out how many healthy person we have in our dataset
pd.read_sql('select count(*) as healthy_person from (select t.Id,sum(t.Calories) as c from(select id,Calories from `fitbit data` where Calories between 2000 and 2500) as t  group by t.Id) as t1',mydb)
```
